## Webscraping 

```python
import requests
from bs4 import BeautifulSoup
import csv


def scrape_and_save_as_csv(url, csv_writer):
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    
    for row in soup.find_all('tr'):  
        data_row = []
        for cell in row.find_all('td'):  
            data_row.append(cell.text.strip())  
        csv_writer.writerow(data_row)


base_url = "https://results.eci.gov.in/PcResultGenJune2024/statewiseU"


csv_filename = 'election_results.csv'


with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)

   
    for i in range(1, 254):
        url = f"{base_url}{i:03}.htm"
        print(f"Scraping {url}")
        scrape_and_save_as_csv(url, csv_writer)
```
